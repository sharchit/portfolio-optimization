```python
selected_columns = ['Close', 'VWAP', 'Volume', 'Turnover', 'Deliverable Volume']
stocks = os.listdir('./data/selected/')
sel = []
for stock in stocks:
    filename = stock.split('.')[0]
    rename_map = {col: f'{filename}_{col}' for col in selected_columns}
    df = pd.read_csv(f'./data/selected/{stock}')
    df = df.rename(mapper=rename_map, axis=1)
    df = df[list(rename_map.values())]
    sel.append(df)
```

```python
master = pd.concat(sel, axis=1)
```

```python
df = pd.read_csv('./data/selected/RELIANCE.csv')
master['Date'] = df.Date
```

```python
master.describe()
```

```python
sel_stocks = os.listdir('./data/selected/')
sel_stocks = [s.split('.')[0] for s in sel_stocks]
log_cols = ['Volume', 'Turnover', 'Deliverable Volume']
```

```python
master2 = master.copy(deep=True)
for company in sel_stocks:
    for col in log_cols:
        master2[f'{company}_{col}'] = master2[[f'{company}_{col}']].apply(lambda x: np.log(x))
```

## Making Action

```python
actions = set()
while len(actions) < 490:
    rand_arrays = np.random.rand(10)
    rand_arrays = rand_arrays / rand_arrays.sum()
    rand_arrays = np.round(rand_arrays/0.05) * 0.05
    if rand_arrays.sum() == 1.0:
        actions.add(tuple(rand_arrays))
```

```python
for i in actions:
    print(i)
    break
```

```python
for i in range(10):
    arr = np.zeros(10)
    arr[i] += 1
    if tuple(arr) not in actions:
        print(i)
        actions.add(tuple(arr))
```

```python
len(actions)
```

```python
action_mat = np.array([np.array(i) for i in actions])
```
